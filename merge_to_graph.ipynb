```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Take in all of our wrestling data and read it into pandas
common = "results/50_most_common.csv"

common_df = pd.read_csv(common)
common_df
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(common_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, common_df["count"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, common_df["word"], rotation="vertical")

# Set x and y limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(common_df["count"])+10)

# Set a Title and labels
plt.title("50 Most Common Words")
plt.xlabel("Word")
plt.ylabel("# of Apperances in Titles")

#print
plt.tight_layout()
plt.savefig("results/50_Most_Common_Words.png")
plt.show()
```
