```python
# Dependencies
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import seaborn as sns
```

```python
# Take in all of our wrestling data and read it into pandas
common_words = "results/50_most_common.csv"

common_words_df = pd.read_csv(common_words)
common_words_df
```

```python
x_axis = np.arange(len(common_words_df))
tick_locations = [value+0.4 for value in x_axis]
plt.figure(figsize=(20,3))
plt.bar(x_axis, common_words_df["count"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, common_words_df["word"], rotation=45)

plt.tight_layout()
plt.savefig("results/50_most_common_words.png")
plt.show()
```

```python
# Take in all of our data and read it into pandas
common = "results/50_most_common_metadata.csv"

common_df = pd.read_csv(common)
common_df
```

```python
text = common_words_df.word.values
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
text
```

```python
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig("results/word_cloud.png")
plt.show()
```

```python
# Filter the DataFrame down only to those columns to chart
ttk = common_df[["title","tempo","key"]]

# Set the index to be "State" so they will be used as labels
ttk = ttk.set_index("title")

ttk
```

```python
# Create a group based on the values in the 'key' column
key_group = ttk.groupby('key')

# Count how many times each key appears in our group
count_keys = key_group['key'].count()

count_keys
```

```python
# Chart our data, give it a title, and la`bel the axes
data_chart = count_keys.plot(kind="bar", title="Songs By Key")
data_chart.set_xlabel("key")
data_chart.set_ylabel("count_key")

plt.savefig("results/songs_by_key.png")
plt.show()
plt.tight_layout()
```
