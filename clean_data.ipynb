```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
file = 'Song-Classification/music.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Preview of the DataFrame
df.head()
```

```python
# Delete extraneous column
del df['artist_mbtags']
df.head()
```

```python
# Identify incomplete rows
df.count()
```

```python
# Delete extraneous column
del df['familiarity']
df.head()
```

```python
# Delete extraneous column
del df['similar']
df.head()
```

```python
# Delete extraneous column
del df['song.hotttnesss']
df.head()
```

```python
df = df.rename(columns={'artist.name': 'artist_name', 'song.id': 'song_id', 'artist.hotttnesss':'artist_hotttnesss'})
```

```python
# Drop all rows with missing information
df = df.dropna(how='any')
```

```python
# Identify incomplete rows
df.count()
```

```python
# The Amount column is the wrong data type. It should be numeric.
df.dtypes
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df.to_csv("music_cleaned.csv", index=False, header=True)
```
